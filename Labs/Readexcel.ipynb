## Check out how to tidy this up to get only what I want out of the spreadsheet

```python
#skip thje first 10 lines
df = pd.read_excel(url,skiprows=10)
```

```python
df.head(10)
```

```python
#drop columns that are not needed
df.dropna(axis=1, how='all', inplace=True)
```

```python
df.head()
```

```python
df.columns
```

```python
df.drop(columns=['CATEGORY (i.e.ISCED description)','R1 Random *','R2 Random*','EOS','EOS Random *','EOS Mid-point','HEI','Test/Interview #','avp','v'],inplace=True)
```

```python
#filter on level 8 courses.
df.loc[df['LEVEL']==8]
```

```python
#write dataframe out to csv file
df.to_csv('cao-points-main/data/cao-2020.csv',sep=',',encoding='utf-8')
```
