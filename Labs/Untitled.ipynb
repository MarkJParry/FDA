```python
import pandas as pd
```

The course titles are inconsistent across institutions so I am setting up some keywords within categories to try and define some disciplines based on the course title, this will aid in groupings later on

```python
#Disciplines and Keywords
Construction=["Construction","Civil Engineering","Architecture","Architectural", "Estate","Agency", "Valuation", "Planning","Quantity"]
Engineering=["Electrical", "Electronic","Food","Agriculture","Mechanical","Chemical","Process","Engineering"]
Computing=["IT","Computer","Information","Software", "Development","Games","Programming","Computing","Analytics","Cloud","Network","Business Information Systems"]
Science=["Science","Biology","Biotechnology","Chemistry","Pharmaceutical","Physics","Astronomy","Mathematics","Forensics"]
Healthcare=["Therapy","Biomedical","Radiography","Nursing","Medicine","Dentistry","Health","Midwifery","Pharmacy","Counselling","Psychology","Physiotherapy"]
Arts_and_Humanities=["Art","Languages","Social","Arts","Cultural","French","German","Russian","Spanish","Italian","Philosophy","Political","Economics","Sociology","English"
,"Music","Film","Television","Stage","TV","Media","Drama"]
Business_and_Law=["Finance","Business","Marketing","Law","Commerce","Accounting","Actuarial","Criminology","Justice"]
Services=["Tourism","Hotel","Catering","Culinary","Bar","Event","Hospitality"]
Education=["Education","Teaching","Coaching"]

#function to examine course title and assign it to a discipline
def discipline(title):
    if any(x in title for x in Construction):
        return "Construction"
    elif any(x in title for x in Engineering):
        return "Engineering"
    elif any(x in title for x in Computing):
        return "Computing"
    elif any(x in title for x in Science):
        return "Science"
    elif any(x in title for x in Healthcare):
        return "Healthcare"
    elif any(x in title for x in Arts_and_Humanities):
        return "Arts_and_Humanities"
    elif any(x in title for x in Business_and_Law):
        return "Business_and_Law"
    elif any(x in title for x in Services):
        return "Services"
    elif any(x in title for x in Education):
        return "Education"
    else:
        return 'Other'
    
#test it out
print(discipline("Tourism and Event Management (Wexford)"))
```

```python
#assign a discipline according to course title and add it as a new column to the dataframe
df = pd.read_csv("data/cao-AllYears.csv") 
df['Discipline'] = df['Course'].apply(discipline)
df.head()
```
